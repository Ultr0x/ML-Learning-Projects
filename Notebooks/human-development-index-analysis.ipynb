# Human Development Index Analysis In R

This will be a short visual representation of the available HDI data in R.

## About Dataset(acc. to author)
HDI is a composite index measuring average achievement in three basic dimensions of human development-a long and healthy life, knowledge and a decent standard of living. The HDI is the geometric mean of normalized indices for each of the three dimensions.

### ***HDI Indicators- Life expectancy at birth, Mean Years of Schooling, Gross national income Per Capita***

## Loading the basic packages. I'll load all other packages as and when needed.

```r
library(tidyverse)
library(dplyr)
library(ggplot2)
```

## Loading the data.

```r
develop<-read.csv("../input/human-development-index-historical-data/HDI.csv")
```

## Overviewing the data.

```r
head(develop)
```

```r
glimpse(develop)
```

## Getting the Top 10 Countries based on HDI Ranking.

```r
develop %>% 
  arrange(HDI.Rank,desc=T) %>% 
  slice(1:10)
```

No surprise to see Iceland,Norway and other Europian Countries....They always leaded the world happiness rating and ranked as top peacful countries as well. ***It would be great to plot these countries with others in a map***

But plotting by already available ranking seems easy and boring...Let's take a long route, will calculate the mean of the HDI Rating of all years and then plot it in the map. By this way we can confirm if the HDI Rating is True or not. Also this will expand the knowledge of variuous other functions in R.

## Feature Engineering.

### changing the ".." values with "NA"

```r
develop[develop==".."]=NA

head(develop)
```

Also, to Calculate means across the rows we need to **change the misentered "Character" colummns to "numeric"**

```r
develop[, c(3:32)] <- sapply(develop[, c(3:32)], as.numeric)

glimpse(develop)
```

Countinuing **Calculating means.**

```r
develop$mean <- rowMeans(subset(develop, select = c(3:32)), na.rm = TRUE)

head(develop$mean)
```

Also I would like to highlight the Top 10 Countries with highest HDI rating on the map along with the color labelling. So, creating a column called **toprank** where it only contains the countries with Highest mean HDI ratings by marking them as ***1*** and rest with ***0***

```r
develop$toprank <- ifelse(develop$HDI.Rank >= 11, 0 & develop$HDI.Rank < 11, 1)

head(develop$toprank,10)
```

## Mapping time :)

For this i'll be using the maps package in R and taking data for whole world map as we r dealing with countries.

```r
library(maps)
mapdata <- map_data("world")

head(mapdata)
```

**Merging the data together to have a combined dataframe**

```r
develop <- rename(develop,c("region" = "Country"))
malmap<-left_join(mapdata,develop,by="region")

head(malmap)
```

Before plotting as i already said that i'll be labelling top 10 countries seperately. To do that we need to pull data from a data frame that specifically contains the top countries alone as this **malmap** dataframe will have all the countires,subregions that will mess up the plot. so ***Creating a subset data frame*** from this one.

```r
# Creating a subset data frame from "malmap"
top10<-malmap %>% 
  filter(toprank=="1") %>% 
  subset(select= -c(8:37))

#Removing duplicates.
top10<-top10 %>%
  arrange(region) %>%
  filter(duplicated(region) == FALSE)

head(top10,10)
```

## Final plotting.

```r
#Loading packages
library(ggrepel)
library(ggthemes)

#Final plotting
ggplot(malmap, aes( x = long, y = lat, group = group )) +
  geom_polygon(aes(color = as.factor(toprank), fill = mean)) +
  scale_color_manual(name = "TOP TEN",values = c('1' = 'black', '0' = NA))+   #Highlighting the top counties with black border.
  scale_fill_gradient(name = "HDI rating", low = "red", high = "green")+      #Highlighting best to worst ratings through red,green gradient.
  guides(colour = "none")+                                                    #Hiding the 1,0 legend
  geom_label_repel(data = top10,aes(label=region),max.overlaps = Inf,size=7)+ #pulling data from top10 dataframe and labelling the top 10 HDI countries.
  theme(panel.grid.major = element_line(color = gray(.5), linetype = "dashed", size = 0.5),
        panel.background = element_rect(fill = "aliceblue"),                  #Customising the plot as a real world map and labelling.
        plot.caption.position = "panel",
        plot.title = element_text(family = "serif",face = "bold",size = 20),
        plot.subtitle = element_text(family="sans",face = "italic",size = 12),
        plot.caption = element_text(family = "mono", face = "italic",size = 15))+
  xlab("Longitude") +
  ylab("Latitude") +
  labs(title = "Map of Human Development Index",
       subtitle = "Labled are top 10 countries with high HDI(excluding HONGKONG)",
       caption = "Indicators that make HDI
       (Life expectancy at birth, Mean Years of Schooling, GNI Per Capita)")
```
